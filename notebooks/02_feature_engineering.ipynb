```python
import pandas as pd
```

```python
df = pd.read_csv("../backend/data/raw/Customer_financial_profiles.csv")
df.head()
```

```python
# (df["yearly_income"] == 0).sum()
df["debt_to_income"] = df["total_debt"] / df["yearly_income"]
df[["total_debt", "yearly_income", "debt_to_income"]].head()
```

```python
df["default_risk"] = ((df["credit_score"]<600) & (df["debt_to_income"]>0.5)).astype(int)
df[["credit_score", "debt_to_income", "default_risk"]].head()
```

```python
print(df["gender"].unique())


df = pd.get_dummies(
    df,
    columns=["gender"],
    dtype=int
)
```

```python
print(df.filter(regex="gender").columns.tolist())
```

```python
columns_to_drop = [
    "id",
    "birth_year",
    "birth_month",
    "transaction_id",
    "date",
    "client_id",
    "card_id",
    "merchant_id",
    "merchant_city",
    "merchant_state",
    "address",
    "zip"
]
df = df.drop(columns=columns_to_drop, errors="ignore")
print(df.columns.tolist())
```

```python
df.to_csv("../backend/data/processed/processed_data.csv", index=False)
print("Processed dataset saved successfully!")
```
