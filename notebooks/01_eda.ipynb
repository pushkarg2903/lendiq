```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Pushkar Garg\OneDrive\Desktop\lendiq\backend\data\raw\Customer_financial_profiles.csv")
```

```python
df.head()
```

```python
print("Dataset Shape:", df.shape)

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="credit_score", bins=30, kde=True)
plt.title("Distribution of credit score")
plt.xlabel("Credit Score")
plt.ylabel("Numer of customers")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="yearly_income", bins=30, kde=True)
plt.title("Distribution of Yearly income")
plt.xlabel("Yearly income")
plt.ylabel("Number of customers")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="yearly_income", y="total_debt")
plt.title("Total Debt vs Yearly Income")
plt.xlabel("Yearly income")
plt.ylabel("Total Debt")
plt.show()
```

```python
avg_credit_score = df.groupby("gender")["credit_score"].mean()
print(avg_credit_score)
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x="gender", y="credit_score", estimator="mean")
plt.title("Average Credit Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Credit Score")
plt.show()
```

```python
numerical_df = df.select_dtypes(include=["number"])
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```
