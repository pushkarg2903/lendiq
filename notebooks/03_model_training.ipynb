```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)

import joblib
```

```python
df = pd.read_csv("../backend/data/processed/processed_data.csv")
df.head()
```

```python
print("Data Type:", df["use_chip"].dtype)
print("Unique Values:", df["use_chip"].unique())
```

```python
df["use_chip"] = df["use_chip"].map({"No": 0, "Yes": 1})

print(df["use_chip"].unique())
df[["use_chip"]].head()
```

```python
x = df.drop(columns=["default_risk", "credit_score", "debt_to_income"])
y = df["default_risk"]
print("Features Shape:", x.shape)
print("Target Shape:", y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print("X_train:", x_train.shape)
print("X_test :", x_test.shape)
print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
```

```python
log_model = LogisticRegression(max_iter=1000, random_state=42)
log_model.fit(x_train, y_train)
print("Logistic Regression model trained successfully!")
```

```python
xgb_model = XGBClassifier(random_state=42, eval_metric="logloss")
xgb_model.fit(x_train, y_train)
print("XGBoost model trained successfully!")
```

```python
log_predictions = log_model.predict(x_test)
xgb_predictions = xgb_model.predict(x_test)
```

```python
print("========== Logistic Regression ==========")

print("Accuracy :", accuracy_score(y_test, log_predictions))
print("Precision:", precision_score(y_test, log_predictions))
print("Recall   :", recall_score(y_test, log_predictions))
print("F1 Score :", f1_score(y_test, log_predictions))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, log_predictions))

print("\nClassification Report:")
print(classification_report(y_test, log_predictions))


print("\n========== XGBoost ==========")

print("Accuracy :", accuracy_score(y_test, xgb_predictions))
print("Precision:", precision_score(y_test, xgb_predictions))
print("Recall   :", recall_score(y_test, xgb_predictions))
print("F1 Score :", f1_score(y_test, xgb_predictions))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, xgb_predictions))

print("\nClassification Report:")
print(classification_report(y_test, xgb_predictions))
```

```python
joblib.dump(xgb_model, "../backend/models/xgboost_model.pkl")
print("XGBoost model saved successfully!")
```
